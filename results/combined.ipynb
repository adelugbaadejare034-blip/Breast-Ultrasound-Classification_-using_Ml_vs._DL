```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt

work_csv =pd.read_csv("work.csv")
```

```python
import matplotlib.pyplot as plt

# Set index to metric name
work_csv.set_index('class', inplace=True)

# Bar chart for accuracy across models
accuracy = work_csv.loc['accuracy'].dropna()

plt.figure(figsize=(8, 5))
plt.bar(accuracy.index, accuracy.values, color='skyblue')
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
# Flatten the values and drop NaNs
values = work_csv.values.flatten()
values = values[~pd.isna(values)]

plt.figure(figsize=(8, 5))
plt.hist(values, bins=5, color='salmon', edgecolor='black')
plt.title('Distribution of All Metric Values')
plt.xlabel('Metric Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Pie chart using f1_score values
f1 = work_csv.loc['f1_score'].dropna()

plt.figure(figsize=(6, 6))
plt.pie(f1.values, labels=f1.index, autopct='%1.1f%%', startangle=140)
plt.title('F1 Score Share by Model')
plt.show()
```
